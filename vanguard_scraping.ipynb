```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.vanguardngr.com/latest-news/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html')
print(soup.prettify())
```

```python
the_soup = soup.find_all('article' , attrs={'class':"entry entry-list-card entry-card-larger"})
print('the length is',len(the_soup))
print(the_soup)
```

```python
the_soup = soup.find_all('img')
print('the length is',len(the_soup))
print(the_soup)

for report in the_soup:
    print(report)
```

```python
# the_soup = the_soup.find_all('img')
print('the length is',len(the_soup))
# print(the_soup)

news = []
img_src  = []

for img in the_soup[0:]:
    alt_text = img.get('alt')
    img_link = img.get('src')

    if alt_text:
        news.append(alt_text)
        img_src.append(img_link)
        print(f'the news {alt_text}')
        print(f'the news image {img_link}')

```

**The Punch News paper**

```python
import requests
from bs4 import BeautifulSoup

url = 'https://punchng.com/all-posts/'
page = requests.get(url)
soup = BeautifulSoup(page.text, 'html.parser')  # Note: using 'html.parser'

news = soup.find_all('article')
for n in news[1:]:
    title = n.h1.a.text.strip()
    link = n.h1.a['href']
    image_link = n.img['data-src']  # Access the 'data-src' attribute

    print('title :',title)
    print('link :', link)
    print('image link :', image_link)

    print(""" """) 
```
